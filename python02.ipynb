```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Iris.csv")
```

# markdown

```python
setosa=df.loc[df['Species']=="Iris-setosa"]
versicolor=df.loc[df['Species']=="Iris-versicolor"]
virginica=df.loc[df['Species']=="Iris-virginica"]
```

```python
plt.plot(setosa['SepalLengthCm'],np.zeros_like(setosa['SepalLengthCm']),'o')
plt.plot(versicolor['SepalLengthCm'],np.zeros_like(versicolor['SepalLengthCm']),'o')
plt.plot(virginica['SepalLengthCm'],np.zeros_like(virginica['SepalLengthCm']),'o')
plt.xlabel("sepallength")
plt.show()
```

# bivarient analysis

```python
sns.FacetGrid(df,hue="Species",size=5).map(plt.scatter,'SepalLengthCm','SepalWidthCm')
plt.show()
```
