# Missing values 

```python
#let us capture the non values
#first let us find the categorical data which is missing
feature_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype=="O"]
for feature in feature_nan:
    print("{}: {}% missing value".format(feature,np.around(dataset[feature].isnull().mean(),4)))
```

```python
#replacing the missing value with the new label or mean
def replace_cat_var(dataset,feature_nan):
    data=dataset.copy()
    data[feature_nan]=data[feature_nan].fillna("missing")
    return data

dataset=replace_cat_var(dataset,feature_nan)
dataset[feature_nan].isnull().sum()
    
    
```

```python
dataset.head()
```

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(dataset[feature].isnull().mean(),4)))
```

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[features for features in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype!="O"]
```

```python
for feature in numerical_with_nan:
    print("{} :{}% missing value".format(feature.np.around(dataset[feature].isnull().mean(),4)))
    
```

```python
#replacing the mising values with median or mode
#we have to replace it with median since there are outliers
for features in numerical_with_nan:
    median_val=dataset[feature].median()
    #create a new features which capturs the nan values
    dataset[features+"nan"]=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_val,inplace=True)
dataset[numerical_with_nan].isnull().sum()
```

```python
for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=dataset[feature].median()
    
    ## create a new feature to capture nan values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[numerical_with_nan].isnull().sum()
```
