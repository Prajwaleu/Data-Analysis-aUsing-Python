# Multicollinearity in linear Regression

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import statsmodels.api as sm
df=pd.read_csv("advertising.csv")
```

```python
df.head()
```

```python
x=df[['TV','Radio','Newspaper']]
y=df[['Sales']]
#we  have to add beta zero value by using ordinary least square(OLS)
```

```python
#ordinarty least square
x=sm.add_constant(x)#adding constant

model=sm.OLS(y,x).fit()
```

```python
x
```

```python
model.summary()
#beta1,beta2,beta3 values are coef to increase the sales we have to incresse the tv
#R-squared error tells about how fit the model is 1 means it fits maximum
```

```python
import matplotlib.pyplot as plt
x.iloc[:,1:].corr()
```

```python
x.iloc[:,:].corr()
```
